```python
import pandas as pd
import numpy as np
```

```python
#task1 load data into dataframe and check for missing values
#drop rows w/ null values 
data = pd.read_csv('Week4PythonDS-ChallengeFIle.csv')
data.shape
```

```python

data_cleaned = data.dropna(how = 'any', axis = 0)
data_cleaned.shape
```

```python
#task2 Thomas, a custy from cali is interested in buying a low priced car made in 2014.
#john needs to identify 100 lowest priced cars from the year 2014 with CA as the state of registration
#create a list of potential cars

pot_cars = data_cleaned[(data_cleaned.state == 'ca') & (data_cleaned.year >= 2014)]

pot_cars.sort_values(by = 'mp', ascending = True)[:100]
```

```python
#task3 thomas is confused by looking at the long list of 100 cars. help john to display the output in a cross
#table based on the car manufactures and prices

pd.crosstab(pot_cars.mp, pot_cars.make, dropna=True)
```

```python
# task4 thomas is interested in chevrolet cars. Eric is offering 10% discount on chevrolet 
# cars from the list of 100 cars identified in the previous task
#display the updated data with only chevrolet cars with their original prices and discount

chevy = pot_cars[pot_cars.make == 'Chevrolet']
chevy
```

```python
discount = chevy['mp'].mul(.9)

discount.rename('mp_discount_10percent',inplace = True)

df_discount = pd.concat([chevy,discount.to_frame()], axis = 1)

df_discount
```

```python
#task5 thomas and several other customers seem to be interested in cars with black or gray colored interiors
#john wants to figure out if the cars with interiors in black or grey are priced higher or lower than cars having interiors of other colors
#determine the average price of all the 100 lowest priced cars from the year 2014 based on the interior color

round(df_discount.groupby(by='interior',as_index = False)['mp'].mean(),2)
```

```python
#task6 most buyers seem to prefer white and black colored cars. john wants to analyze the color-wise 
# availability of cars for each manufacturer from the years 2010-2015
ext_color = data_cleaned[(data_cleaned.year >= 2010) & (data_cleaned.year <= 2015)]
groupby_ext_color =ext_color.groupby(by = ['make','color'])['color'].count()

df = pd.DataFrame(groupby_ext_color)
df
```

```python
#task 7
#Eric, another customer is interested in a good condition BMW Sedan manufactured between 2010 and 2015.
#John wants to display the details of all such cars that have a condition score >=4.5

# help john figure out the year wise maximum expected priced of the bmw cars 
#manufactured between 2010 and 2015 grouped by each model

bmw = data_cleaned[(data_cleaned.body  == 'Sedan') & (data_cleaned.make == 'BMW') & (data_cleaned.year >= 2010) & (data_cleaned.year <= 2015) & (data_cleaned.condition >= 4.5)]
bmw_model = pd.DataFrame(bmw.groupby(['model'])['mp'].max())
bmw_model
```

```python
#task 8 
#john wants to recommend Eric to even Consider SUV's from other manufacturers before finalizing the car
# john needs to list the top 100 SUV's available between the years 2010 and 2015, based on their prices
# in descending order with an odometer reading <= 30000

SUV = data_cleaned[(data_cleaned.body  == 'SUV') & (data_cleaned.year >= 2010) & (data_cleaned.year <= 2015) & (data_cleaned.odometer <= 30000)]

SUV.sort_values(by = 'mp', ascending = False)[:100]
```
