```python
import pandas as pd
import numpy as np
```

```python
#task1 Load the data from the file into a DataFrame and identify the missing values (if any) 

data = pd.read_csv('Week4PythonDS_practiceFile.csv')



data.isnull().sum() #this does a 'group by' by columns and adds all those that are null
```

```python
#Treat the missing values with the most frequently found data in that column.

#The DataFrame['column_name'].value_counts() command will help you identify the most 
# frequently found data for any column with missing values.

data['test1'].value_counts() #takes the column and finds which values were used the most

mean_test1 = round(data['test1'].mean(),2)
median_test2 = round(data['test2'].median(),2)

# .fillna() fills in any n/a null values
data['test1'] = data['test1'].fillna(mean_test1)
data['test2'] = data['test2'].fillna(median_test2)

data.isnull().sum()
```

```python
#task 2 Identify and display the duplicate rows (if any). Delete the duplicate rows in the data.
print(data.shape)
dropped = data.drop_duplicates()
print(dropped.shape)
data
```

```python
#task3 In the column named "tuitionfee_waiver", replace the value "Does not qualify" and "Qualifies 
# for reduced/free" with No and Yes, respectively. Display the updated column data and verify if the values are replaced.
#data['tuitionfee_wavier'] 

replaced_data = data.replace(to_replace=['Does not qualify','Qualifies for reduced/free '], value=['No','Yes'])
replaced_data
```

```python
#task4 Calculate and display the average of test1 scores of students at each school.
#Display the name of the school that has the highest average test1 score.
#Hint: Use the "group by" operation to first segregate the data for each school and then calculate the average test score.

school = round(data.groupby(by=['school'], as_index = False)['test1'].mean(),2)
print(school)
      
sorted_school = school.sort_values(by = ['test1'],ascending = False)


value = sorted_school['school'][16]
highest_score = sorted_school['test1'][16]#this is index number 16

print(f'\n The school {value} had the highest average test1 score of {highest_score} \n')


print(sorted_school[:1])#this is the first row
```

```python
#task5 Display the highest test2 score of each school in a tabular format.
#Display the name of the top five schools based on the output of the above task.

school2 = data.groupby(by=['school'], as_index = False)['test2'].max()

school2.sort_values(by = 'test2', ascending = False)[:5]# only taking the top 5
```

```python
# task5 different method-using a pivot table

Pvt = pd.pivot_table(data, index=['school'], values=['test2'],aggfunc='max')
Pvt.sort_values(by = 'test2', ascending = False)[:5]
```
